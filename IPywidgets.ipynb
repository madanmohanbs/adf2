```python
import ipywidgets
import IPython.display
```

```python
x1 = ipywidgets.Label(value="Label")
x2 = ipywidgets.Text(placeholder="Text")
x3 = ipywidgets.DatePicker(description='Date..')
x4 = ipywidgets.ToggleButtons(options=['AA','BB'])
x5 = ipywidgets.Button(description='Button')
x6 = ipywidgets.IntSlider(value=3, max=7,min=2)
```

```python
output = ipywidgets.Output()
def on_button_clicked(event):
  with output:
    IPython.display.clear_output()
    print(f"x1={x1.value}")
    print(f"x2={x2.value}")
    print(f"x3={x3.value}")
    print(f"x4={x4.value}")
    print(f"x6={x6.value}")
    
x5.on_click(on_button_clicked)
```

```python
box = ipywidgets.VBox([x1,x2,x3,x4,x5,x6,output])
display(box)
```
